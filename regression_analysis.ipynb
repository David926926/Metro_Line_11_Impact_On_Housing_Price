```python
import numpy as np
import statsmodels.api as sm
import pandas as pd

file_path = "/Users/zhuxiaowen/Desktop/24 Fall/Econometrics 1/Project/test_12.13.xlsx"  
sheet_name = "Data_for_test7"  
data = pd.read_excel(file_path, sheet_name=sheet_name)

# Preprocess the data
# Create the dependent variable: Log of transaction price
data['log_price'] = np.log(data['Transaction price'])

# Define independent variables
# Distance dummies
distance_vars = ['0-400', '400-800', '800-1200', '1200-1600', '1600-2000']

# Control variables
control_vars = [
    'House Area', 
    'Quantity of bedrooms', 
    'Floor',
    "Decoration",
    'Orientation',
    'Age'
]

# Interaction terms (BEFORE * Distance dummies)
for var in distance_vars:
    interaction_term = f'POST_{var}'
    data[interaction_term] = data['POST'] * data[var]

interaction_vars = [f'POST_{var}' for var in distance_vars]

# Combine all independent variables
independent_vars = distance_vars + control_vars + interaction_vars

# Add a constant term for the regression
X = sm.add_constant(data[independent_vars])

# Dependent variable
y = data['log_price']

# Fit the regression model
model = sm.OLS(y, X).fit(cov_type="HC1")

# Display the regression results
print(model.summary())
```
